**boxplot**

```python
df8 = sns.load_dataset("tips")
df8
```

```python
sns.set(style= "whitegrid")

sns.boxplot(x = "day", y = "total_bill", hue = "sex", data= df8)
plt.show()
```

```python
sns.set(style= "whitegrid")

sns.boxplot(x = "day", y = "total_bill", hue = "sex", data= df8 , order = ["Fri", "Sun", "Thur", "Sat"])
plt.show()
```

```python
sns.set(style= "whitegrid")

sns.boxplot(x = "day", y = "total_bill", showmeans = True , color = "yellow" ,data= df8, 
           meanprops = {"marker": "o", "markeredgecolor":  "r", "markersize" :15}, 
                       linewidth = 3,  fliersize = 5, palette = 'plasma')
plt.show()
```

```python
sns.set(style= "whitegrid")

sns.boxplot( color = "yellow" ,data= df8, 
           meanprops = {"marker": "o", "markeredgecolor":  "r", "markersize" :15}, 
                       linewidth = 3,  fliersize = 5, palette = 'plasma', orient = "h")
plt.show()
```

```python
sns.set(style= "whitegrid")

sns.boxplot(y=df8["total_bill"])
plt.show()
```
